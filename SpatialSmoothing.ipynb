# Spatial regularisation example

This notebook contains a simple spatial regularisation example, similar to the "temporal" smoothing example shown in [NDVI_smoothing.ipynb](http://nbviewer.ipython.org/url/raw.github.com/jgomezdans/toy_eoldas/master/NDVI_smoothing.ipynb). I will make use of some useful functions from the EO-LDAS user's guide (in particular, see [here](http://jgomezdans.github.com/eoldas_release/example_spatial.html#spatial-data-assimilation)).

```python
def gauss_kern(size, sizey=None):
    """ 
    Returns a normalized 2D gauss kernel array for convolutions 
    From http://www.scipy.org/Cookbook/SignalSmooth
    """
    import numpy as np
    size = int(size)
    if not sizey:
        sizey = size
    else:
        sizey = int(sizey)
    x, y = np.mgrid[-size:size+1, -sizey:sizey+1]
    g = np.exp(-(x**2/float(size)+y**2/float(sizey)))
    return g / g.sum()

def create_data ( nr,nc,n_per=1, noise=0.15, obs_off=0.33, \
					window_size=3):
    """
    Create synthetic "NDVI-like" data for a fictitious space series. We return
    the original data, noisy data (using IID Gaussian noise), the QA flag as well
    as the time axis.
    
    Missing observations are simulated by drawing a random number between 0 and 1
    and checking against obs_off.
    
    Parameters
    ----------
    nc : number of columns
    nr : number of rows

    n_per : integer
    Observation periodicity. By default, assumes every sample

    noise : float
    The noise standard deviation. By default, 0.15

    obs_off : float
    The threshold to decide on missing observations in the series.

    
    """
    import numpy as np
    from scipy import signal

    r,c = np.mgrid[1:nr+1:n_per,1:nc+1:n_per]
    ndvi_clean =  np.clip(np.sin(np.pi*2*(r-1)/nr)*np.sin(np.pi*2*(c-1)/nc), 0,1) 
    ndvi = ndvi_clean.copy()
    # add Gaussian noise of sd noise
    ndvi = np.random.normal(ndvi,noise,ndvi.shape)
     
    # set the qa flags for each sample to 1 (good data)
    qa_flag = np.ones_like ( ndvi).astype( np.int32 )
    passer = np.random.rand(ndvi.shape[0],ndvi.shape[1])
    if window_size >0:
        # force odd
        window_size = 2*(window_size/2)+1
        # smooth passer
        g = gauss_kern(window_size)
        passer = signal.convolve(passer,g,mode='same')
    # assign a proportion of the qa to 0 from an ordering of the smoothed 
    # random numbers
    qf = qa_flag.flatten()
    qf[np.argsort(passer,axis=None)[:passer.size * obs_off]]  = 0
    qa_flag = qf.reshape(passer.shape)
    return ( r,c , ndvi_clean, ndvi, qa_flag )

```

The two functions above allow us to create a spatial array, add noise to it, and flick some pixels off to simulate the effect of cloud cover. We are using again a univariate series of for example NDVI. We can see how the data and the truth look like by creating some data with `create_data` (usurprisingly!) and plotting it. Note that the missing observations are shown in grey:

```python
rows = 40
cols = 40
sigma_obs = 0.15
rows, cols, ndvi_true, ndvi_obs, qa_flag = create_data ( rows, cols )

cmap=plt.cm.spectral
cmap.set_bad('0.8')
plt.subplot ( 1,2,1 )
plt.imshow ( ndvi_true, interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.title("'True'")
plt.subplot ( 1,2,2)
plt.title("Observed")
plt.imshow ( np.ma.array(ndvi_obs,mask=qa_flag==0), \
    interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
```

From `NDVI_smooothing.ipynb`, we know what the assumptions are:

* Everything is Gaussian
* Everything is linear

As before, we shall use an identity operator to map from state space to observation, so think of the problem at hand as trying to solve for the "true field of NDVI" from a noisy and incomplete observation, but helping ourselves from the idea that the field should be spatially correlated, and smooth. For the time being, we shall ignore having a prior distribution on the values of NDVI. Thus, we define the following two elements of our cost function:

1. The "fit to the observations" (where these are available)
$$
J_{obs} =-\frac{1}{2}\sum_{i}^{N_{rows}}\sum_{j}^{N_{cols}} \frac{qa_{i,j}\cdot(x_{i,j} - x^{o}_{i,j})^2}{\sigma_o^2}
$$

2. The fit to the process model (or spatial smoothness expectation):
$$
J_{smooth} = -\frac{1}{2}\sum_{i}^{N_{rows}}\sum_{j}^{N_{cols}} \sum_{ii=-1}^{1}\sum_{jj=-1}^{1} \frac{(x_{i,j} - x^{}_{i+ii,j+jj})^2}{\sigma_m^2}
$$

Wee see that we introduce $qa_{i,j}$, the value of the QA flag (or mask) for pixel $i,j$. The second term is a way of writing the smoothness constraint without matrices, just as a number of embedded summations. It basically means that for each pixel $i,j$ we consider its eight neighbours: $(i-1,j-1)\,(i,j-1)\,(i+1,j-1)\,(i-1,j)\,(i+1,j)\,(i-1,j+1)\,(i-1,j+1)\,(i-1,j+1)$. We assume that the smoothness term is isotropic and that it affects all directions equally (we ought to take into account the fact that the diagonal neighbours are a bit further away though!)

```python
def fit_obs ( x, obs, sigma_obs, qa ):
    """
    A fit to the observations term. This function returns the likelihood

    
    """
    der_j_obs = np.where ( qa_flag == 1, (x - obs)/sigma_obs**2, 0 )
    j_obs = np.where ( qa_flag == 1, 0.5*(x - obs)**2/sigma_obs**2, 0 )
    j_obs = j_obs.sum()
    return ( j_obs, der_j_obs )

print "Examples"
print "========"
print "J_{obs} for true",  fit_obs ( ndvi_true, ndvi_obs, 0.15, qa_flag )[0]
print "J_{obs} for noise",  fit_obs ( ndvi_obs, ndvi_obs, 0.15, qa_flag )[0]
```

So, we see that if we feed in the true data, we have a fairly high value of $J_{obs}$, as it does not match the noisy observations. On the other hand, if we fill in the observations, we get a perfect fit. Note that $J_{obs}$ is only calculated where there are observations (i.e., `qa_flag == 1`), so the value in the gaps is undetermined from this function.

```python
def fit_smoothness (  x, sigma_model  ):
    """
    This function calculates the spatial smoothness constraint. We use
    a numpy strides trick to efficiently evaluate the neighbours. Note
    that there are no edges here, we just ignore the first and last
    rows and columns. Also note that we could have different weights
    for x and y smoothing (indeed, diagonal smoothing) quite simply
    """
    # Build up the 8-neighbours
    hood = np.array ( [  x[:-2, :-2], x[:-2, 1:-1], x[ :-2, 2: ], \
                    x[ 1:-1,:-2], x[1:-1, 2:], \
                    x[ 2:,:-2], x[ 2:, 1:-1], x[ 2:, 2:] ] )
    j_model = 0
    der_j_model = x*0
    for i in [1,3,4,6]:#range(8):
        j_model = j_model + 0.5*np.sum ( ( hood[i,:,:] - x[1:-1,1:-1] )**2 )/sigma_model**2
        der_j_model[1:-1,1:-1] = der_j_model[1:-1,1:-1] - ( hood[i,:,:] - x[1:-1,1:-1] )/sigma_model**2
    return ( j_model, 2*der_j_model )
    
print "Examples"
print "========"
print "J_{smooth} for noise",  fit_smoothness ( ndvi_obs, .1 )[0]
print "J_{smooth} for true",  fit_smoothness ( ndvi_true, .1 )[0]
```

In this case, the noisy observations have a large value of $J_{smooth}$ due to the high frequency variability introduced by the random noise. The "true field" on the other hand has a much smaller value, but still not quite 0: there are sharp edge transitions in the region between the hills and the flatlands. Regularisation will tend to smooth over these sharp interfaces.

## Testing the derivatives

It's easy to make mistakes calculating the derivative, so it's useful to test it against a numerical approximation:


```python
import scipy.optimize
def the_deriv ( x, obs, sigma_obs, sigma_model, qa ):
    x = x.reshape( obs.shape )
    ( j_obs, der_j_obs ) = fit_obs ( x, obs, sigma_obs, qa )
    ( j_model, der_j_model ) = fit_smoothness ( x, sigma_model )
    return (der_j_obs + der_j_model).flatten() 

def test_cost ( x, obs, sigma_obs, sigma_model, qa ):
    x = x.reshape( obs.shape )
    ( j_obs, der_j_obs ) = fit_obs ( x, obs, sigma_obs, qa )
    ( j_model, der_j_model ) = fit_smoothness ( x, sigma_model )
    return j_obs + j_model

for i in xrange(10):
    x = np.random.rand(40,40)
    approx_der = scipy.optimize.approx_fprime( x.reshape(1600), test_cost, 1e-10, ndvi_obs, 0.15, 0.1, qa_flag)
    deriv = the_deriv(x.reshape(1600), ndvi_obs, 0.15, 0.1, qa_flag)
    plt.plot ( approx_der, deriv, 'k,')
plt.plot([-800,800],[-800,800], '-r' )
plt.xlabel("Numerical approximation")
plt.ylabel("Analytic derivative")
```

# Solving the problem

We now have the cost function as well as the derivatives (in fact, it's all a bit sub-optimal in the way it's put together, but there you go!). We can now use `scipy.optimize` and minimize the cost function. One thing to remember is the optimisers requires vectors of parameters, so we need to flatten the 2D arrays when passing them through. 

The following code snipped defined a cost function that returns the cost and its partial derivatives, and then loops over some values of $\gamma$. We also calculate the $\ell_2$-norm of the difference between predicted field and "true" field, and plot it as a function of $\sigma=1/\gamma$. As $\sigma$ goes to 0 (too much regularisation), we get an oversmoothed reconstruction. Converseley, if $\sigma \rightarrow \infty$, we get too little smoothness (the "process model" is ignored). The optimal value is found to be around 0.23 in this particular experiment.

Note that although visually, values of $\sigma=0.18$ appear "better", the system is oversmoothing, and probably blurring the sharp transitions between the "yolks" and the rest of the scene.

```python
import scipy.optimize
sigma_model = 1./10.
sigma_obs = 0.15
fito={}
def the_fit ( x, obs, sigma_obs, sigma_model, qa ):
    x = x.reshape( obs.shape )
    ( j_obs, der_j_obs ) = fit_obs ( x, obs, sigma_obs, qa )
    ( j_model, der_j_model ) = fit_smoothness ( x, sigma_model )
    #print j_model, j_obs
    return ( j_obs + j_model, (der_j_obs + der_j_model).flatten() )
plt.subplots_adjust ( hspace=0.3)
for (i,sigma_model) in enumerate(np.logspace(-2,0,20)):
    x0 = np.random.rand  ( ndvi_obs.flatten().shape[0])*0
    retval = scipy.optimize.fmin_l_bfgs_b( the_fit, x0, \
                args=( ndvi_obs, sigma_obs, sigma_model, qa_flag), \
                iprint=1, disp=1)
    plt.subplot(5,4,i+1)
    plt.imshow ( retval[0].reshape((40,40)), interpolation='nearest', vmin=0,vmax=1, cmap=plt.cm.spectral)
    plt.title( r'$\sigma=$'+"%5.2G"%sigma_model, fontsize=6)
    plt.xticks(visible=False)
    plt.yticks(visible=False)
    fito[i] = np.linalg.norm ( ndvi_true - retval[0].reshape((40,40)))
plt.figure();plt.plot(np.logspace(-2,0,20), fito.values(), 'o-')  
plt.xlabel(r'$1/\gamma$')
plt.ylabel("Difference to Truth")
plt.axvline ( np.logspace(-2,0,20)[np.argmin(fito.values())], c='k')
xarrow = np.logspace(-2,0,20)[np.argmin(fito.values())]
yarrow = fito[np.argmin(fito.values())]
ax = plt.gca()
ax.annotate(r'$\sigma=$'+"%5.2G"%xarrow, xy=(xarrow, yarrow), xytext=(0.4, 8),
            bbox=dict(boxstyle="round", fc="0.8"),
            arrowprops=dict(arrowstyle="->"))
            
```

See how setting $\sigma=0.16\,$ results on the marked edges around the centre. Values close to the optimum don't suffer from this: the residual starts to look a bit more like random noise.

```python
x0 = np.random.rand  ( ndvi_obs.flatten().shape[0])*0
retval = scipy.optimize.fmin_l_bfgs_b( the_fit, x0, \
                args=( ndvi_obs, sigma_obs, 0.16, qa_flag), \
                iprint=1, disp=1)
plt.subplot(1,2,1)
plt.imshow ( np.abs(ndvi_true - retval[0].reshape(ndvi_true.shape)), interpolation='nearest', vmin=0, vmax=0.4, cmap=plt.cm.spectral)
plt.xticks(visible=False)
plt.yticks(visible=False)
x0 = np.random.rand  ( ndvi_obs.flatten().shape[0])*0
retval = scipy.optimize.fmin_l_bfgs_b( the_fit, x0, \
                args=( ndvi_obs, sigma_obs, 0.23, qa_flag), \
                iprint=1, disp=1)
plt.subplot(1,2,2)
plt.imshow ( np.abs(ndvi_true - retval[0].reshape(ndvi_true.shape)), interpolation='nearest', vmin=0, vmax=0.4, cmap=plt.cm.spectral)
plt.xticks(visible=False)
plt.yticks(visible=False)
```

### How fast is this?

It's important to assess the speed of the processing. In the next snippet, we show that solving a particular instance of the problem is quite fast: it requires around 50-100 ms for a $40\times 40 $ problem. It takes less than 2s for a $200\times 200$ problem.

```python
import scipy.optimize
import timeit
sigma_model = 0.18
def the_fit ( x, obs, sigma_obs, sigma_model, qa ):
    x = x.reshape( obs.shape )
    ( j_obs, der_j_obs ) = fit_obs ( x, obs, sigma_obs, qa )
    ( j_model, der_j_model ) = fit_smoothness ( x, sigma_model )
    #print j_model, j_obs
    return ( j_obs + j_model, (der_j_obs + der_j_model).flatten() )


def the_funco (x0 = np.random.rand  ( ndvi_obs.flatten().shape[0])*0 ):
    retval = scipy.optimize.fmin_l_bfgs_b( the_fit, x0, \
                args=( ndvi_obs, sigma_obs, sigma_model, qa_flag), \
                iprint=1, disp=1)
    return retval

%%timeit the_funco()
```

## Multi-resolution

An annoying fact of EO data spatial resolutions are often different, spanning from the few meters to the kilometer. In some senarios, moderate resolution data provides very frequent acquisitions, which allows the possibility of capturing the land surface when clouds are persistent. Higher resolution data, due to typicall smaller swaths, usually has fewer revisit opportunities and that can result in few actual observations. Additionally, some high resolution data might suffer from poor spectral or angular sampling. In any case, we want to combine both datasets in an optimal way to make full use of the information contained in both observations. 

```python
rows = 200
cols = 200
sigma_obs = 0.15
rows, cols, ndvi_true_hires, ndvi_obs_hires, qa_flag_hires = create_data ( rows, cols, obs_off=0.6, noise=0.20, window_size=11 )

cmap=plt.cm.spectral
cmap.set_bad('0.8')
plt.subplot ( 2,2,1 )
plt.imshow ( ndvi_true_hires, interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.ylabel("High Resolution")
plt.title("'True'")
plt.subplot ( 2,2,2)
plt.title("Observed")
plt.imshow ( np.ma.array(ndvi_obs_hires,mask=qa_flag_hires==0), \
    interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.ylabel("$\sigma=0.25$\nObsOff=60%%")
rows = 100
cols = 100
sigma_obs = 0.15
rows, cols, ndvi_true_lores, ndvi_obs_lores, qa_flag_lores = create_data ( rows, cols, noise=0.15,obs_off=0.3)

cmap=plt.cm.spectral
cmap.set_bad('0.8')
plt.subplot ( 2,2,3 )
plt.imshow ( ndvi_true_lores, interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)


plt.ylabel("Low Resolution")
plt.subplot ( 2,2,4)
plt.imshow ( np.ma.array(ndvi_obs_lores,mask=qa_flag_lores==0), \
    interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.ylabel("$\sigma=0.15$\nObsOff=20%%")
cax = plt.axes([0.01, 0.05, 0.85, 0.05])
plt.colorbar(cax=cax,orientation='horizontal')
```

In the previous snippet, we have created two datasets: a high and a low resolution dataset. The high resolution dataset is sparser (more missing data), and has a higher noise contribution. The lower resolution dataset has more data (ie you could have two MODIS overpasses on the same day, but unlikely to have to TM overpasses). In the previous Bayesian scheme, what we have is two different "windows" of observing the state $\mathbf{x}$. For convenience, we shall assume that $\mathbf{x}$ is a representation of the state on the high resolution grid (we could choose any other grid, but this simplifies things). The mapping from state to high resolution observation is performed by an identity operator, whereas the lowe resolution observations are treated as an average value of $\matbhf{x}$ under the low resolution footprint: $\mathbf{B}\cdot\mathbf{x}$. Note that in this case, we can also accommodate different IFOVs by constructing $\mathbf{B}$ differently.

Note that the smoothness constraint term does not vary, only the fit to the observations needs to change. In order to quickly do the matrix summation product, we produce a downsampling function.

```python
def downsample(myarr,factor):
    """
    Downsample a 2D array by averaging over *factor* pixels in each axis.
    Crops upper edge if the shape is not a multiple of factor.

    This code is pure numpy and should be fast.
    Leeched from <https://code.google.com/p/agpy/source/browse/trunk/agpy/downsample.py?r=114>
    """
    xs,ys = myarr.shape
    crarr = myarr[:xs-(xs % int(factor)),:ys-(ys % int(factor))]
    dsarr = numpy.concatenate([[crarr[i::factor,j::factor] 
        for i in range(factor)] 
        for j in range(factor)]).mean(axis=0)
    return dsarr
a = downsample(ndvi_true_hires, 2 )
plt.subplot(1,2,1)
plt.imshow ( np.ma.array(ndvi_true_lores,mask=qa_flag_lores==0), \
    interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.subplot(1,2,2)
plt.imshow ( a, \
    interpolation='nearest', vmin=0, vmax=1, cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
```

So, let us define the new "fit to the observation" functions. The first one, for the high resolution data is the same that we had used before, but I repeat it here with a slightly different name, `fit_obs_dir`. For the low resolution data, I use `fit_obs_spat`. The only new thing here is the paramteer `factor`, which is purely the factor between the spatial resolutions. The function first "downsamples" $\mathbf{x}$ to the appropriate resolution and uses the downsampled version of the state, `xa` to calculate the likelihood. The derivative is also calculated, but it is then interpolated to be the same size as the original state. 

```python
def fit_obs_dir ( x, obs, sigma_obs, qa_flag ):
    """
    A fit to the observations term. This function returns the likelihood

    
    """
    #print x.shape, obs.shape, sigma_obs, qa_flag.shape
    der_j_obs = np.where ( qa_flag == 1, (x - obs)/sigma_obs**2, 0 )
    j_obs = np.where ( qa_flag == 1, 0.5*(x - obs)**2/sigma_obs**2, 0 )
    j_obs = j_obs.sum()
    return ( j_obs, der_j_obs )

def fit_obs_spat ( x, obs, sigma_obs, qa_flag, factor ):
    """
    A fit to the observations term. This function returns the likelihood

    
    """
    import scipy.ndimage.interpolation
    
    xa = downsample ( x, factor )
    xa = xa
    assert ( xa.shape == obs.shape )
    der_j_obs = np.where ( qa_flag == 1, (xa - obs)/sigma_obs**2, 0 )
    
    der_j_obs = scipy.ndimage.interpolation.zoom ( der_j_obs, factor, order=1 )
    j_obs = np.where ( qa_flag == 1, 0.5*(xa - obs)**2/sigma_obs**2, 0 )
    j_obs = j_obs.sum()
    return ( j_obs, der_j_obs )


def the_fit ( x, obs_hires, obs_lores, sigma_obs_hires, sigma_obs_lores, sigma_model, qa_hires, qa_lores, factor=2 ):
    
    x = x.reshape( obs_hires.shape )
    
    ( j_obs_hires, der_j_obs_hires ) = fit_obs_dir ( x, obs_hires, sigma_obs_hires, qa_hires )
    
    ( j_obs_lores, der_j_obs_lores ) = fit_obs_spat ( x, obs_lores, sigma_obs_lores, qa_lores, factor=factor )
    #j_obs_lores = 0.0
    #der_j_obs_lores = der_j_obs_hires*0.
    
    ( j_model, der_j_model ) = fit_smoothness ( x, sigma_model )
    #print j_model, j_obs
    the_derivative = np.ascontiguousarray ( (der_j_obs_hires + der_j_obs_lores  + der_j_model).flatten() )
    
    return ( j_obs_hires + j_obs_lores + j_model), the_derivative
           

```

Let's just try solving the problem. Let's set the uncertainties to match the values we had when we created the data above, and use a "reasonable" value of $\sigma_{model}$. We shall then use `scipy.optimize.fmin_l_bfgs_b` to minimise the cost function. Note that the function to minimise, `the_fit`, is defined just above this paragraph. It uses the previously defined fit to the model as well as the high- and low-resolution fit to the observation terms (as well as their derivatives).

```python

sigma_model = 0.018
sigma_obs_hires = 0.2
sigma_obs_lores = 0.15

x0 = ndvi_true_hires.flatten()*0.
retval = scipy.optimize.fmin_l_bfgs_b( the_fit, x0, \
                args=( ndvi_obs_hires, ndvi_obs_lores, sigma_obs_hires, sigma_obs_lores, sigma_model, qa_flag_hires, qa_flag_lores), iprint=1, disp=1)
```

It's useful to see the retrieved state (it's MAP value) and compare it with the synthetic truth:

```python
r=retval[0]
plt.subplot(1,2,1)
plt.title("Retrieved MAP")
plt.imshow ( r.reshape(ndvi_obs_hires.shape) ,vmin=0,vmax=1, interpolation='nearest',cmap=cmap )
plt.xticks(visible=False)
plt.yticks(visible=False)
plt.subplot(1,2,2)
plt.title("'Truth'")
plt.imshow ( ndvi_true_hires, interpolation='nearest', vmin=0, vmax=1, cmap=cmap)
plt.xticks(visible=False)
plt.yticks(visible=False)

cax = plt.axes([0.1, 0.1, 0.85, 0.075])
plt.colorbar(cax=cax,orientation='horizontal')
```

The above plot shows some interesting effects: while the reconstruction is reasonable and very good (compare to the originals!!), it is also easy to see the signal "bleeding" from the two fried eggs, in contrast with the sharp edges in the original. Also, the central peaks in the "yolk" are smoothed over. This is an inevitable result of using a smoothness constraint: it does achieve smoothness but there are situations where we don't actually want to have smoothness. 

We will also perform the experiment we carried out earlier on to estimate the optimal value of $\sigma_{model}$. This exercise is only possible because we can compare the result of the assimilation/smoothing/inference (delete as appropriate) with the simulated truth. Also, we are not taking into account the full predictive ability of the model: we just calculate some simple RMSE term that does not take into account how certain we are about the prediction. This could clearly be done.

## Dealing with uncertainty

So far so good, but we haven't actually explored how to deal with uncertainty. In our current set-up, and under the assumptions of Gaussianity and linearity (or wekly-nonlinearity), the posterior distribution is Gaussian. We've seen already how we can calculate the mode of the posterior, and because for a normal distribution the model and the mean are the same, we only need to estimate the covariance structure of the posterior. This is well-known and given by the inverse of the Hessian at the posterior mode. One issue of interest is that the Hessian will likely be quite large (in the previous example, it will be a $40\,000\times 40\,000\;$ matrix). Its calculation will take a long time, and its inverse will also be slow. We note however that the Hessian is made up of several contributions:

* a prior (if we were using it. It was ignored in the previous discussion),
* a likelihood (or several) term,
* a "model"/"smoothness" term.

These matrices are usually large, but very sparse (for example, the prior, assuming independence is purely diagonal, and the model contribution to the Hessian matrix will only have a couple of diagonals which are non-zero). The likelihood terms will be zero or block-constant in many cases.
